```python
# # training the model through dataset
from ultralytics import YOLO

# Load the model.
model = YOLO('yolov8x.pt')  #yolov8 contains total 80 classes such as person, pen, etc

# Training.
results = model.train(
   data='/content/pothole_dataset_v8/pothole.yaml',  #providing yaml file of dataset it contains the path of training and validating images and the no. of classes
   imgsz=1280,
   epochs=100,
   batch=8,
)
```
